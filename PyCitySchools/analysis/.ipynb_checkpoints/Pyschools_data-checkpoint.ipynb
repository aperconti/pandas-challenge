# Data Findings


### Number of students: 
The data shows a strong correlation to charter schools having a higher overall passing rate than public school districts. 

### Size of schools: 
After consulting the Pyschools data, there is a clear trend showing higher student numbers in public school districts than in Charter schools.

### Reading vs. Math scores: 
In this data set, the average reading scores are higher than average math scores across all school sizes. This trend continues based on spending per student, showing a higher passing rate in reading regardless of higher per-student spending and vice versa. 

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_loaded = "../Resources/School.csv"
student_data_loaded = "../Resources/Students.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_loaded)
student_data = pd.read_csv(student_data_loaded)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Calculate the total number of schools
school_count = school_data_complete["School ID"].nunique()
print(f'Total Schools: {school_count}')

#Calculate the total number of students
student_count = school_data_complete["Student ID"].nunique()
print(f'Total Students: {student_count}')

# #Calculate the total budget
budgets = school_data_complete.loc[school_data_complete["Student ID"]]["budget"]
total_budget = budgets.unique().sum()

#applying currency function to budget column in 'school_data_complete'
school_data_complete['budget'] = school_data_complete['budget'] #.apply(currency_format)

#Calculate the average math score
math_score = school_data_complete["math_score"]
math_average = math_score.mean()
total_math_av = round(math_average,6)
print(f'Average Math Score: {total_math_av}')

#Calculate the average reading score
reading_score = school_data_complete["reading_score"]
reading_score = reading_score.mean()
total_reading_av = round(reading_score, 6)
print(f'Average Reading Score: {total_reading_av}')


#Calculate the overall passing rate 
overall_av_score = (total_reading_av + total_math_av) / 2
overall_av_score_r = round(overall_av_score,6)
print(f'% Overall Passing Rate: {overall_av_score_r}')

#Passing math score (70 or greater)
math_passing = school_data_complete["math_score"] >= 70
true_count = sum(math_passing)
passing_math_score = round((true_count/student_count)*100,6)
print(f'% Passing Math: {passing_math_score}')

#passing reading score (70 or greater)
reading_passing = school_data_complete["reading_score"] >= 70
true_count_reading = sum(reading_passing)
passing_reading_score = round((true_count_reading/student_count)*100,6)
print(f'% Passing Reading: {passing_reading_score}')

#creating a dataframe 
#input variables 
district_summary = [school_count, student_count,total_budget, total_math_av, overall_av_score_r, total_reading_av, passing_math_score, passing_reading_score] 
#dataframe columns
df = pd.DataFrame([district_summary], columns =['Total Schools','Total Students','Total Budget', 'Average Math Score', 'Average Reading Score','% Passing Math', '% Passing Reading', '% Overall Passing Rate']) 

#show dataframe
school_data_complete

```

```python
# Overview table

# Get unique rows
grouped_overview_table = school_data_complete.drop_duplicates(subset = ["school_name"])

# Remove unused columns
unused_columns = ['Student ID', 'math_score', 'student_name', 'gender', 'grade', 'reading_score']
grouped_overview_table = grouped_overview_table.drop(unused_columns, axis=1)

# Assign School Name as Index
grouped_overview_table = grouped_overview_table.drop(['School ID'], axis=1)

# Rename column names
grouped_overview_table.columns = ['School Name', 'School Type', 'Total Students', 'Total School Budget']
grouped_overview_table = grouped_overview_table.reset_index(drop=True)

# Create Per Student Budget
grouped_overview_table['Per Student Budget'] = grouped_overview_table['Total School Budget'] / grouped_overview_table['Total Students']

# Create Average Math Score
student_mean_math_df = student_data.groupby('school_name')['math_score'].mean()
#grouped_overview_table = grouped_overview_table.merge(student_mean_math_df, left_on='School Name', right_on='school_name')


""" 
===========================================
START Create Average Reading Score
===========================================
"""
student_mean_reading_df = student_data.groupby('school_name')['math_score'].mean()
grouped_overview_table = grouped_overview_table.merge(student_mean_reading_df, left_on='School Name', right_on='school_name')
grouped_overview_table.columns = ['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score']
""" 
===========================================
END Create Average Reading Score
===========================================
"""


""" 
===========================================
START Create Average Reading Score
===========================================
"""
student_mean_reading_df = student_data.groupby('school_name')['reading_score'].mean()
grouped_overview_table = grouped_overview_table.merge(student_mean_reading_df, left_on='School Name', right_on='school_name')
grouped_overview_table.columns = ['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score']
""" 
===========================================
END Create Average Reading Score
===========================================
"""

""" 
===========================================
START Creation of the % Passing Math Column
===========================================
"""
# Get table of total number of students in each school (size of school)
student_size_math_df = student_data.groupby('school_name')['math_score'].size()

# Get table that meet passing criteria (students who scored 70 or above in math)
student_mean_math_df = student_data.groupby('school_name')['math_score'].apply(lambda x: x >= 70)

# Count the number of students who passed, grouped by each school
student_math_count = student_data[student_mean_math_df].groupby('school_name')['math_score'].count()

# Divide number of students who passed by total number of students, grouped by school
raw_percentage_passing = student_math_count.div(student_size_math_df) * 100

# Add results to the overview table
grouped_overview_table = grouped_overview_table.merge(raw_percentage_passing, left_on='School Name', right_on='school_name')

# Clean up unused columns
grouped_overview_table.columns = ['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math']
""" 
=========================================
END Creation of the % Passing Math Column
=========================================
"""

""" 
==============================================
START Creation of the % Passing Reading Column
==============================================
"""
# Get table of total number of students in each school (size of school)
student_size_reading_df = student_data.groupby('school_name')['reading_score'].size()
# Get table that meet passing criteria (students who scored 70 or above in math)
student_mean_reading_df = student_data.groupby('school_name')['reading_score'].apply(lambda x: x >= 70)

# Count the number of students who passed, grouped by each school
student_reading_count = student_data[student_mean_reading_df].groupby('school_name')['reading_score'].count()

# Divide number of students who passed by total number of students, grouped by school
raw_percentage_passing = student_reading_count.div(student_size_reading_df) * 100

# Add results to the overview table
grouped_overview_table = grouped_overview_table.merge(raw_percentage_passing, left_on='School Name', right_on='school_name')

# Clean up unused columns
grouped_overview_table.columns = ['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading']

""" 
============================================
END Creation of the % Passing Reading Column
============================================
"""

""" 
==============================================
START Creation of the Overall Passing Rate (Average of the above two)
==============================================
"""
grouped_overview_table["% Overall Passing Rate"] = ( grouped_overview_table["% Passing Reading"] + grouped_overview_table["% Passing Math"] ) / 2
""" 
============================================
END Creation of the % Passing Reading Column
============================================
"""
#reset index
grouped_overview_table.set_index('School Name', inplace=True)

grouped_overview_table
```
